## Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.graphics.tsaplots as sgt 
import statsmodels.tsa.stattools as sts 
from statsmodels.tsa.seasonal import seasonal_decompose
import yfinance
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
import warnings
warnings.filterwarnings("ignore")
sns.set()
```

### Loading Finance Data

```python
raw_data = yfinance.download (tickers = "DLF.NS", start = "2015-01-01", 
                              end = "2022-08-23", interval = "1d", group_by = 'ticker', auto_adjust = True, treads = True)
```

```python
raw_data
```

### Data Transformation

```python
df = pd.DataFrame()
df = raw_data[['Close']]
df=df.asfreq('b')
df.fillna(method='ffill',inplace=True)
```

### Data Exploration

```python
plt.figure(figsize=(20,8))
plt.plot(df)
plt.ylabel('Price in INR')
plt.xlabel('Date')
plt.title('Price (High) of Infosys Stock',fontsize=(16))
```

## Seasonality

```python
s_dec_additive = seasonal_decompose(df.Close, model = "additive")
s_dec_additive.plot()
plt.show()
```

## Autocorrelation

```python
sgt.plot_acf(df.Close, lags = 60, zero = False)
plt.title("ACF Power", size = 20)
plt.show()
```

## Partial Auto Correlation

## Data Preprocessing

```python
#Splitting Data into Train and Test
size = int(len(df)*0.85)
train,test = df[:size],df[size:]
```

```python
# Feature Scaling
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
```

```python
train_scaled = sc.fit_transform(train)
test_scaled = sc.fit_transform(test)
```

```python
# Creating a train data structure with 50 timesteps and 1 output
X_train = []
y_train = []
for i in range(50, len(train)):
    X_train.append(train_scaled[i-50:i, 0])
    y_train.append(train_scaled[i, 0])
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
# Creating a test data structure with 50 timesteps and 1 output
X_test = []
y_test = []
for i in range(50, len(test)):
    X_test.append(test_scaled[i-50:i, 0])
    y_test.append(test_scaled[i, 0])
X_test, y_test = np.array(X_test), np.array(y_test)
```

```python
# Reshaping
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

## DEEP NEURAL NETWORK

```python
net = Sequential()
net.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 1)))
net.add(Dropout(0.2))
net.add(LSTM(units = 50, return_sequences = True))
net.add(Dropout(0.2))
net.add(LSTM(units = 50, return_sequences = True))
net.add(Dropout(0.2))
net.add(LSTM(units = 50))
net.add(Dropout(0.2))
net.add(Dense(units = 1))
```

```python
net.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
# Fitting the Neural Network
net.fit(X_train, y_train, epochs = 150, batch_size = 100)
```

```python
## Predicting Prices
preds = net.predict(X_test)
predicted_price = sc.inverse_transform(preds)
```

```python
df_2 = test[50:]
df_2['Predicted_Close'] = predicted_price
```

## Result

```python
plt.figure(figsize=(20,8))
df_2['Close'].plot()
df_2['Predicted_Close'].plot()
plt.legend(['Real Price','Predicted Price'])
plt.title('DLF STOCK REAL AND PREDICTED PRICE',fontsize=18)
plt.ylabel('Price in INR',fontsize=13)
plt.xlabel('Date',fontsize=13);
```
